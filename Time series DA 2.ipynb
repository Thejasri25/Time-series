#### <center>QUESTION 1</center>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa. stattools import adfuller
from numpy import log
```

```python
train=pd.read_csv("D:/VIT LECTURE NOTES/TIME SERIES LAB/Microsoft stock price.csv")
train
```

```python
plt.plot(train['Close'])
```

```python
# checking the stationarity.
plt.plot(train['Close'])
result1 = adfuller(train['Close'].dropna())
print('ADF Statistic : %f ' % result1[0])
print('p-value : %f ' % result1[1])
```

### we can see that the data is non stationary

```python
# for the p-value we can see that the data is non stationary .
# now using trasformations to change non stationry to stationary.
#log Transformation
x=np.log(train['Close'].dropna())
plt.plot(x)
result = adfuller(x)
print('ADF Statistic : %f ' % result[0])
print('p-value : %f ' % result[1])
```

```python
#power transformation
y=np.log(train['Close'])
y['diff']=y-y.shift(1)
y['diff'].dropna().plot()
result2=adfuller(y['diff'].dropna())
print('ADF Statistic : %f ' % result2[0])
print('p-value : %f ' % result2[1])
```

### Conclusion:

#### As we observe above initially the data is non-stationary and further by applying the diffrencing techniques such as Log transformation and Power transformation the data follows stationarity.This particular data becomes stationary after power transformation only
