#### <center>QUESTION 1</center>
##### Calculating MA for the passengers data

##### Input

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa. stattools import adfuller
from numpy import log
```

```python
train=pd.read_csv("D:/VIT LECTURE NOTES/TIME SERIES LAB\Air passengers.csv")
train
```

#### Output:

##### moving average 3rd order

```python
train['SMA_3']=train['Passengers'].rolling(window=3).mean()
print(train['SMA_3'].tail(30))
train['Passengers'].plot(color='green',lw=3,figsize=(12,6))
train['SMA_3'].plot(color='red',lw=3,figsize=(12,6))
plt.title('The air passengers travelling data',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Number of passengers')
```

```python
x=train['SMA_3'].dropna()
plt.plot(x)
result=adfuller(x)
print('ADF Stat: %f'%result[0])
print('P value: %f'%result[1])
```

#### Moving average 5th order

```python
train['SMA_5']=train['Passengers'].rolling(window=5).mean()
print(train['SMA_5'].head(30))
```

```python
train['Passengers'].plot(color='green',lw=3,figsize=(12,6))
train['SMA_5'].plot(color='red',lw=3,figsize=(12,6))
plt.title('The air passengers travelling data',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Number of passengers')
```

```python
x=train['SMA_5'].dropna()
plt.plot(x)
result=adfuller(x)
print('ADF Stat: %f'%result[0])
print('P value: %f'%result[1])
```

#### Moving average 7th order

```python
train['SMA_7']=train['Passengers'].rolling(window=7).mean()
print(train['SMA_7'].tail(30))
train['Passengers'].plot(color='green',lw=3,figsize=(12,6))
train['SMA_7'].plot(color='red',lw=3,figsize=(12,6))
plt.title('The air passengers travelling data',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Number of passengers')
```

```python
x=train['SMA_7'].dropna()
plt.plot(x)
result=adfuller(x)
print('ADF Stat: %f'%result[0])
print('P value: %f'%result[1])
```

#### Moving averaage 10th order

```python
train['SMA_10']=train['Passengers'].rolling(window=10).mean()
print(train['SMA_10'].tail(30))
train['Passengers'].plot(color='green',lw=3,figsize=(12,6))
train['SMA_10'].plot(color='red',lw=3,figsize=(12,6))
plt.title('The air passengers travelling data',fontsize=20)
plt.xlabel('Data')
plt.ylabel('Number of passengers')
```

```python
x=train['SMA_10'].dropna()
plt.plot(x)
result=adfuller(x)
print('ADF Stat: %f'%result[0])
print('P value: %f'%result[1])
```

### Conclusion:

#### We can observe that there is too much variation in the simple moving averages for the give data.

#### The ADF test values for the MA 3rd order and 10th order are similar and also 5th order as well as 7th order moving averages differ by 0.2
