#### <center>Null Hypothesis(H0):The product value is non stationary </center>
#### <center>Alternate Hypothesis(H1):The product value is stationary </center>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

#### Input: 

```python
datavalue=pd.read_csv("D:/VIT LECTURE NOTES/TIME SERIES LAB/Dataset1.csv")
datavalue.head()
```

```python
datavalue.info()
```

```python
datavalue.describe()
```

```python
plt.figure(figsize=(15,10))
plt.plot(datavalue.Value,color='orange')
plt.show()
```

```python
#find the shape of the data to split into two halves
datavalue.shape
```

```python
#split the data into two 
datavalue1=datavalue[:199]
datavalue2=datavalue[199:]
```

```python
#check for stationarity by comparing means of two groups of the data
print(datavalue1.Value.mean())
print(datavalue2.Value.mean())
```

```python
#check for stationarity by comparing Standard deviation of two groups of the data
print(datavalue1.Value.std())
print(datavalue2.Value.std())
```

#### output: 

```python
#performing Ad fuller test
t1=datavalue.Value
result1=adfuller(t1)
print('ADF statistics: ',result1[0])
print('p-value: ',result1[1])
```

### Conclusion:
#### If p value is greater than significance level, we reject null hypothesis. 
#### Here, p value is greater than significance level of 0.05, so we reject null hypothesis
#### Therefore, our data is stationary
